```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
from LogisticRegression import LogisticRegression

```

```python
# Set a random seed for reproducibility
np.random.seed(42)

# Cell 2: Data Loading and Preparation (use the same split as regression)
iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42, stratify=y
)

```

```python
# Variant 1 - Petal Length/Width
print("\n--- Classifier 1: Petal Features ---")
X_train1 = X_train[:, 2:] # Petal length/width
X_test1 = X_test[:, 2:]

# We need a copy of the model instance for plotting
class_model1 = LogisticRegression(learning_rate=0.1, n_iterations=3000)
class_model1.fit(X_train1, y_train)
accuracy1 = class_model1.score(X_test1, y_test)
print(f"Accuracy with Petal features: {accuracy1:.2f}")

# Visualize
plt.figure(figsize=(8, 5))
plot_decision_regions(X_train1, y_train, clf=class_model1)
plt.title("Decision Regions: Petal Features")
plt.xlabel("Petal Length [cm]")
plt.ylabel("Petal Width [cm]")
plt.savefig("classifier1_regions.png")
plt.show()

```

```python
# Variant 2 - Sepal Length/Width
print("\n--- Classifier 2: Sepal Features ---")
X_train2 = X_train[:, :2] # Sepal length/width
X_test2 = X_test[:, :2]

class_model2 = LogisticRegression(learning_rate=0.1, n_iterations=3000)
class_model2.fit(X_train2, y_train)
accuracy2 = class_model2.score(X_test2, y_test)
print(f"Accuracy with Sepal features: {accuracy2:.2f}")

# Visualize
plt.figure(figsize=(8, 5))
plot_decision_regions(X_train2, y_train, clf=class_model2)
plt.title("Decision Regions: Sepal Features")
plt.xlabel("Sepal Length [cm]")
plt.ylabel("Sepal Width [cm]")
plt.savefig("classifier2_regions.png")
plt.show()

```

```python
# Variant 3 - All Features
print("\n--- Classifier 3: All Features ---")
# Use the full dataset
class_model3 = LogisticRegression(learning_rate=0.1, n_iterations=3000)
class_model3.fit(X_train, y_train)
accuracy3 = class_model3.score(X_test, y_test)
print(f"Accuracy with all features: {accuracy3:.2f}")
print("\n(Visualization is not possible for 4 features)")
```
